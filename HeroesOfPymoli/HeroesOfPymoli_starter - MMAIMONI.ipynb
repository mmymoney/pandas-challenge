```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
#Count unique players using nunique function
total_players = purchase_data["SN"].nunique()

#print total number of players
print(f' The total number of players is {total_players}.')
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Compute Unique Item Count, Average Price, Total Purchases, and Total Revenue

unique_items = purchase_data["Item Name"].nunique()
unique_items
avg_price = purchase_data['Price'].mean()
avg_price
total_ct_purchases = purchase_data["Purchase ID"].count()
total_ct_purchases
total_revenue = purchase_data["Price"].sum()
total_revenue

#Create Summary Data Frame

dict = {'Unique Items':unique_items, 'Average Price':avg_price, 'Total Purchases':total_ct_purchases,'Total Revenue':total_revenue}
df=pd.DataFrame(dict, index=[0])

#Clean up formatting
df["Average Price"] = df["Average Price"].map('${:,.2f}'.format)
df["Total Revenue"] = df["Total Revenue"].map('${:,.2f}'.format)

#display summary dataframe
df
```

## Gender Demographics

* Percentage and Count of Other / Non-Disclosed




```python
#Remove duplicates and calculate total number of players in each category (leveraging total player count calculated above)

players_deduped = purchase_data.drop_duplicates(subset="SN",keep="first")
gender_demo = players_deduped["Gender"].value_counts()

#create dataframe from value counts
gender_demodf = pd.DataFrame(gender_demo)

#retitle column 'Gender' to 'Count of Gender'
gender_demodf["Count of Gender"] = gender_demodf["Gender"]

#calculate percentage of total
gender_demodf["Percentage of Total"] = gender_demodf["Count of Gender"]/total_players

#remove 'Gender' column
del gender_demodf["Gender"]

#clean up formatting
gender_demodf["Percentage of Total"] = gender_demodf["Percentage of Total"].map('{:,.2%}'.format)

#display summary dataframe
gender_demodf

```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Create data frames for each of the three 'Gender' classifications
male_purchases = purchase_data.loc[purchase_data["Gender"] == "Male",:]
female_purchases = purchase_data.loc[purchase_data["Gender"] == "Female",:]
other_purchases = purchase_data.loc[purchase_data["Gender"] == "Other / Non-Disclosed",:]

# Calculate Total Purchases by Gender
male_purchases_count = len(male_purchases["Price"])
female_purchases_count = len(female_purchases["Price"])
other_purchases_count = len(other_purchases["Price"])

# Calculate Total Purchase Value by Gender
male_purchases_total = male_purchases["Price"].sum()
female_purchases_total = female_purchases["Price"].sum()
other_purchases_total = other_purchases["Price"].sum()

# Calculate Average Purchase Price by Gender
total_purchase = purchase_data["Price"].sum()
avg_male_purchases = male_purchases_total/len(male_purchases["SN"])
avg_female_purchases = female_purchases_total/len(female_purchases["SN"])
avg_other_purchases = other_purchases_total/len(other_purchases["SN"])

# Calculate Average Per Person Purchase Total by Gender (dropping duplicate SNs)

avg_maleperson_purchases = male_purchases_total / len((male_purchases.drop_duplicates(subset="SN",keep="first")))
avg_femaleperson_purchases = female_purchases_total / len((female_purchases.drop_duplicates(subset="SN",keep="first")))
avg_otherperson_purchases = other_purchases_total / len((other_purchases.drop_duplicates(subset="SN",keep="first")))

#Create list of tuples to populate summary dataframe using calculated variables
purchases_genders = [(male_purchases_count, avg_male_purchases, male_purchases_total, avg_maleperson_purchases),
                    (female_purchases_count, avg_female_purchases, female_purchases_total, avg_femaleperson_purchases),
                    (other_purchases_count, avg_other_purchases, other_purchases_total, avg_otherperson_purchases)]

#create and populate summary data frame, adding index label
gender_purchasesdf = pd.DataFrame(purchases_genders, columns = ['Purchase Count', 'Average Purchase Price', 'Total Purchase Value', 'Avg Total Purchase per Person'], index=['Male', 'Female', 'Other/Non-Disclosed'])
gender_purchasesdf.index.name = "Gender"

#clean up formatting
gender_purchasesdf["Average Purchase Price"] = gender_purchasesdf["Average Purchase Price"].map('${:,.2f}'.format)
gender_purchasesdf["Total Purchase Value"] = gender_purchasesdf["Total Purchase Value"].map('${:,.2f}'.format)
gender_purchasesdf["Avg Total Purchase per Person"] = gender_purchasesdf["Avg Total Purchase per Person"].map('${:,.2f}'.format)

#display summary dataframe
gender_purchasesdf
```

```python
#Create bins and labels
bins = [0,9,14,19,24,29,34,39,100]
labels = ["<10","10-14", "15-19","20-24","25-29","30-34","35-39","40+"]

#form binned data using purchase data and form dataframe
binned_data1 = pd.cut(purchase_data["Age"],bins=bins, labels = labels)
binned_datadf1 = pd.DataFrame(binned_data1)

#Add back 'Price' column for calculations
binned_datadf1["Price"] = purchase_data["Price"]

#Calculate Total Revenue by Age Bin
revenue_df = binned_datadf1.loc[binned_datadf1['Age'] == "<10","Price"].sum()


#Generate counts by bin and form dataframe
summary_binned_data1 = binned_datadf1["Age"].value_counts()
summary_binned_datadf1 = pd.DataFrame(summary_binned_data1)
summary_binned_datadf1["Purchase Count"] = summary_binned_datadf1["Age"]

#form binned data using purchase data (removing duplicate SNs to identify unique players) and form dataframe
binned_data2 = pd.cut(players_deduped["Age"],bins=bins, labels = labels)
binned_datadf2 = pd.DataFrame(binned_data2)
summary_binned_data2 = binned_datadf2["Age"].value_counts()
summary_binned_datadf2 = pd.DataFrame(summary_binned_data2)

#Create dataframe with total purchase values computed using conditional loc's
data = {'Age':['20-24','15-19','25-29','30-34','35-39','10-14','<10','40+'],'Total Purchase Value':[binned_datadf1.loc[binned_datadf1["Age"] == "20-24","Price"].sum(),binned_datadf1.loc[binned_datadf1["Age"] == "15-19","Price"].sum(),binned_datadf1.loc[binned_datadf1["Age"] == "25-29","Price"].sum(),binned_datadf1.loc[binned_datadf1["Age"] == "30-34","Price"].sum(),binned_datadf1.loc[binned_datadf1["Age"] == "35-39","Price"].sum(),binned_datadf1.loc[binned_datadf1["Age"] == "10-14","Price"].sum(),binned_datadf1.loc[binned_datadf1["Age"] == "<10","Price"].sum(),binned_datadf1.loc[binned_datadf1["Age"] == "40+","Price"].sum()]}
total_average = pd.DataFrame(data)
total_average.set_index(total_average['Age'],drop=True,inplace=True)


summary_binned_datadf1["Total Purchase Value"] = total_average["Total Purchase Value"]
del summary_binned_datadf1["Age"]   

#Calculate Average Purchase Price based on Purchase Count and Total Purchase Value
summary_binned_datadf1["Average Purchase Price"] = summary_binned_datadf1["Total Purchase Value"]/summary_binned_datadf1["Purchase Count"]
summary_binned_datadf1["Player Count"] = summary_binned_datadf2["Age"]

#Calculate Average Total Purchase per person using Total Purchase Value and Player Count (removing duplicate SNs)
summary_binned_datadf1["Avg Total Purchase per Person"] = summary_binned_datadf1["Total Purchase Value"]/summary_binned_datadf1["Player Count"]

#delete Player Count column as this is unwanted in final output
del summary_binned_datadf1["Player Count"]

#Set index name
summary_binned_datadf1.index.name = "Age Ranges"

#clean up formatting
summary_binned_datadf1["Average Purchase Price"] = summary_binned_datadf1["Average Purchase Price"].map('${:,.2f}'.format)
summary_binned_datadf1["Total Purchase Value"] = summary_binned_datadf1["Total Purchase Value"].map('${:,.2f}'.format)
summary_binned_datadf1["Avg Total Purchase per Person"] = summary_binned_datadf1["Avg Total Purchase per Person"].map('${:,.2f}'.format)

#display summary data frame
summary_binned_datadf1
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Obtain instances of purchases by each player and store in dataframe
playercounts = purchase_data["SN"].value_counts()
playercountsdf = pd.DataFrame(playercounts)

#Group and sum purchases by player and store in dataframe
playerspend = purchase_data.groupby('SN')["Price"].sum()
playerspenddf = pd.DataFrame(playerspend)

#Sort (descending) and rename 'Price' column to Total Purchase Value 
playerspenddf_sorted = playerspenddf.sort_values(by=["Price"],ascending=False)
playerspenddf_sorted["Total Purchase Value"] = playerspenddf_sorted["Price"]

#Add Purchase Count column to summary dataframe
playerspenddf_sorted["Purchase Count"] = playercountsdf["SN"]

#Create column with calculated value for average purchase price
playerspenddf_sorted["Average Purchase Price"] = playerspenddf_sorted["Total Purchase Value"]/playerspenddf_sorted["Purchase Count"]

#clean up formatting
playerspenddf_sorted["Average Purchase Price"] = playerspenddf_sorted["Average Purchase Price"].map('${:,.2f}'.format)
playerspenddf_sorted["Total Purchase Value"] = playerspenddf_sorted["Total Purchase Value"].map('${:,.2f}'.format)

#drop redundant column
del playerspenddf_sorted["Price"]

#display summary dataframe (header 5)
playerspenddf_sorted.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#groupby item ID and sum Price column to derive 'Total Purchase Value' by item
item_id_grouped = purchase_data.groupby('Item ID')["Price"].sum()
item_id_groupeddf = pd.DataFrame(item_id_grouped)
item_id_groupeddf

#create dataframe to derive purchase count by item
item_id_count = purchase_data["Item ID"].value_counts()
item_id_countdf = pd.DataFrame(item_id_count)
item_id_countdf.sort_index()

#create dataframe to derive item price per unit (removing duplicate Item IDs)
item_price = purchase_data.drop_duplicates(subset="Item ID",keep="first")
item_pricedf = pd.DataFrame(item_price)
item_pricedf_sorted = item_pricedf.sort_values(by='Item ID',ascending=True)
item_pricedf_new = item_pricedf_sorted.set_index('Item ID')

#Add supplemental columns to the dataframe (Item Name, Purchase Count, Item Price); rename Price column to Total Purchase Value
item_id_groupeddf["Purchase Count"] = item_id_countdf["Item ID"]
item_id_groupeddf["Total Purchase Value"] = item_id_groupeddf["Price"]
item_id_groupeddf["Item Price"] = item_pricedf_new["Price"]
item_id_groupeddf["Item Name"] = item_pricedf_new["Item Name"]

#sort descending purchase count
item_id_groupeddf_sorted = item_id_groupeddf.sort_values(by="Purchase Count",ascending=False)

#remove redundant column
del item_id_groupeddf_sorted["Price"]

#store separate dataframe for Most Profitable Item problem
item_id_groupeddf_sorted1 = item_id_groupeddf_sorted
item_id_groupeddf_sorted1 = item_id_groupeddf_sorted1[['Item Name','Purchase Count','Item Price','Total Purchase Value']]



#Reorder columns
item_id_groupeddf_sorted = item_id_groupeddf_sorted[['Item Name','Purchase Count','Item Price','Total Purchase Value']]

#clean up formatting
item_id_groupeddf_sorted["Item Price"] = item_id_groupeddf_sorted["Item Price"].map('${:,.2f}'.format)
item_id_groupeddf_sorted["Total Purchase Value"] = item_id_groupeddf_sorted["Total Purchase Value"].map('${:,.2f}'.format)


#display summary dataframe (header 5)
item_id_groupeddf_sorted.head()
```
